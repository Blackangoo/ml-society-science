```python
from randomised_response import rp_binary
```

```python
n_people = 1000
n_features = 1

drug_use = 0.01
import numpy as np
data = np.random.choice(2, size=[n_people, n_features], p = [1 - drug_use, drug_use])
```

```python
error = abs(np.mean(data) - drug_use)
print(error)
```

```python
import matplotlib.pyplot as plt
epsilon_range = np.linspace(0,0.5)
errors = [abs(np.mean(rp_binary(data, epsilon))) for epsilon in epsilon_range ]
```

```python
plt.plot(epsilon_range, errors)
```
