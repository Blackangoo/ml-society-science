```python
from randomised_response import rp_binary
```

```python
n_people = 100000
n_features = 1

drug_use = 0.1
import numpy as np
data = np.random.choice(2, size=[n_people, n_features], p = [1 - drug_use, drug_use])
```

```python
error = abs(np.mean(data) - drug_use)
print(error)
```

```python
import matplotlib.pyplot as plt
epsilon_range = np.linspace(0,10)

errors_all = [abs(drug_use - np.mean(rp_binary(data, epsilon))) for epsilon in epsilon_range ]
errors_10 = [abs(drug_use - np.mean(rp_binary(data[1:10000], epsilon))) for epsilon in epsilon_range ]
errors_100 = [abs(drug_use - np.mean(rp_binary(data[1:1000], epsilon))) for epsilon in epsilon_range ]
errors_1000 = [abs(drug_use - np.mean(rp_binary(data[1:100], epsilon))) for epsilon in epsilon_range ]
errors_10000 = [abs(drug_use - np.mean(rp_binary(data[1:10], epsilon))) for epsilon in epsilon_range ]
```

```python
plt.plot(epsilon_range, errors_all)
plt.plot(epsilon_range, errors_10)
plt.plot(epsilon_range, errors_100)
plt.plot(epsilon_range, errors_1000)
plt.plot(epsilon_range, errors_10000)
plt.legend(["1", "1/10", "1/100", "1/1000", "1/10000"])
```
