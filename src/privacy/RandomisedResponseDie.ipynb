```python
v = 0.1

def posterior_randomised_response(prior, v, answer):
    if (answer==True):
        return prior * v / (prior * v + (1 - v)*(1 - prior))
    else:
        return (1 - v) * prior / ((1 - v)*prior + v * (1 - prior))

```

```python
## for fixed mechanism, vary the prior, and see the posterior
import numpy as np
prior = np.linspace(0,1)
```

```python
v=.1
posterior_yes = posterior_randomised_response(prior, v, True)
posterior_no = posterior_randomised_response(prior, v, False)
```

```python
import matplotlib.pyplot as plt
```

```python
# The effect of the prior on learning learning the truth
# (it has none)
plt.plot(prior, posterior_yes)
plt.plot(prior, posterior_no)
```

```python
prior = 0.1
v = np.linspace(0,0.5)
posterior_yes = posterior_randomised_response(prior, v, True)
posterior_no = posterior_randomised_response(prior, v, False)
```

```python
# The effect of the mechanism's randomness on learning the truth
plt.plot(v, posterior_yes)
plt.plot(v, posterior_no)
```
